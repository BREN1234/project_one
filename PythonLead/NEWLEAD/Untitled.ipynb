```python
import requests,re,datetime
```

```python
from bs4 import BeautifulSoup
```

```python
urll= "https://www.indeed.com.sg/jobs?q=java+developer&l=Singapore&radius=25&sort=date&limit=50&start=1&fromage=1"
soup= BeautifulSoup(requests.get(urll).text,'lxml')
numberOfHits= soup.find("div",id="searchCount").text
print("Numebr of hits",numberOfHits)
numberOfHits= [re.sub(r'[^A-Za-z0-9]+', '', x) for x in numberOfHits.strip().split(" ")]
print("Numebr of hits",numberOfHits[-2])
if len(numberOfHits) == 6:
    numberOfJobs=int(numberOfHits[5])
if len(numberOfHits) == 5:
    numberOfJobs=int(numberOfHits[3])
```

```python
records= soup.find_all("div",class_="jobsearch-SerpJobCard unifiedRow row result")
```

```python
len(records)
```

```python
for record in records:
    print("Designation: ",str(record.find('a',class_="jobtitle turnstileLink ").text).strip())
    print("Company: ",str(record.find("span",class_='company').text).strip())
    print("Location: ",str(record.find("span",class_="location").text).strip())
    print("Summary: ",str(record.find("div",class_="summary").text).strip())
    print("Date: ",str(record.find("span",class_='date').text).strip())
    dt= None
    l=str(record.find("span",class_='date').text).strip()
    li=re.sub(r'[^0-9A-Z]','',l)
    try:
        if li.isalpha():
            dt=datetime.datetime.today().date()
        else:
            dt=(datetime.datetime.today()-datetime.timedelta(days=int(li))).date()
    except ValueError as e:
        print("i am out")
        print(e)
    print(dt)
    print(len(record.find('span',{"class":"company"}).findChildren()))
```
